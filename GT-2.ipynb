```python
import numpy as np
import matplotlib.pyplot as plt
from astropy import units as u


a = 7.56e-16 * u.J / u.m**3 / u.K**4 # Stefan-Boltzmann constant in W/m^2/K^4
K_b = 1.380649e-23 * u.J / u.K  # Boltzmann constant in J/K
G = 6.67430e-11 * u.N * u.m**2/ (u.kg **2)  # Gravitational constant in m^3/(kg*s^2)
M_sun = 1.989e30 * u.kg  # Mass of the Sun in kg
m_h = 1.6735575e-27 * u.kg  # Mass of hydrogen

A = 18 * K_b**4 / (np.pi * a * (G**3) * (M_sun**2) * (m_h**4))
print(f"valor de A = {A}")

def factor_masas(miu, x):

    return np.sqrt(A*x/(miu**4*(1-x)**4))

x = np.linspace(0, 0.85, 100)

# Graficar la curva para =0.62, la metalicidad solar
y_2 = factor_masas(0.62, x)
plt.plot(y_2, x, label="=0.62", color="red", linestyle="-.")

# graficar para las demas metalicidades 
for miu in np.linspace(0.5, 2, 11):
    y = factor_masas(miu, x)
    plt.plot(y, x, label=f"={miu:.2f}")


plt.axhline(0.5, color="blue", lw=0.8, linestyle="--", label=r"$\beta =0.5$")
plt.axvline(1, color="black", lw=1.2, linestyle=":", label=r"$M_\odot=1$")
plt.xlabel(r"$( M*/ M_\odot)$")
plt.ylabel(r"$1- \beta_*$")
plt.xlim(0, 50)
plt.grid()
plt.legend(loc="center left", bbox_to_anchor=(1, 0.5))
```
